# Exploratory Data Analysis (EDA) on a Movies Dataset

```python
#1. importing the libraries and completing setup

# i)pandas for working with the dataframes
import pandas as pd
# ii)matplotlib for the plotting of the data 
import matplotlib.pyplot as plt
# ii) numpy for dealing with any calculations
import numpy as np
# iii)reading the csv file and converting it into a data frame
movies = pd.read_csv("IMDb_Dataset.csv",index_col = 0)
```

```python
# 2. Initial Exploratory Data Analysis (EDA)
#   i)the basic view of our dataframe and top 5 data entries
movies.head()
```

```python
#   ii) all the columns and their datatypes and entry's information
movies.info()
```

```python
#   iii) the number of non-null values, The arithmetic mean of the values, The standard deviation, a measure of data dispersion, The minimum value, first quartile, median, third quartile, The maximum value.
movies.describe()
```

```python
#   iv) the rows and column information
print("Number of rows in the dataframe:",movies.shape[0])
print("Number of Columns in the dataframe:",movies.shape[1])
```

```python
# v)chexking for any missing value and filling it with 0 just incase
if(movies.isna().sum().sum()!=0):
    movies.fillna(0)
movies.isna().sum()
```

```python
#   vi) top rated movies
top_rated = movies.sort_values("IMDb Rating", ascending=False)
top_rated = top_rated.drop_duplicates(subset="IMDb Rating")
top_rated.head(10)
```

```python
# visualizing top rated movies
top_10_B = top_rated.head(10)
plt.figure(figsize=(10,5))
plt.barh(top_10_B.index,top_10_B["IMDb Rating"],color="skyblue")
plt.xticks(np.arange(0, 10.5, 0.5))
plt.xlabel("IMDb Rating")
plt.ylabel("Movie Title")
plt.title("Top 10 Movies")
plt.show()
```

```python
#   vii) worst rated movies
worst_rated = movies.sort_values("IMDb Rating")
worst_rated=worst_rated.drop_duplicates(subset="IMDb Rating")
worst_rated.head(10)
```

```python
# visualizing worst rated movies
top_10_W = worst_rated.head(10)
plt.figure(figsize=(10,5))
plt.barh(top_10_W.index,top_10_W["IMDb Rating"],color="skyblue")
plt.xticks(np.arange(0, 7, 0.5))
plt.xlabel("IMDb Rating")
plt.ylabel("Movie Title")
plt.title("Top 10 Worst Movies")
plt.show()
```

```python
#   viii)longest movies
longest = movies.sort_values("Duration (minutes)",ascending=False)
longest.head()
```

```python
# visualizing longest movies 
_10_longest = longest.head(10)
plt.figure(figsize=(10,6))
plt.barh(_10_longest.index, _10_longest["Duration (minutes)"], color="lightcoral")
plt.xlabel("Runtime (minutes)")
plt.ylabel("Movie Title")
plt.title("Top 10 Longest Movies")
plt.gca().invert_yaxis()  
plt.xticks(np.arange(0, _10_longest["Duration (minutes)"].max() + 30, 10)) 
plt.show()
```

```python
#   viii)shortest movies
shortest = movies.sort_values("Duration (minutes)")
shortest.head()
```

```python
# visualizing shortest movies 
_10_shortest = shortest.head(10)
plt.figure(figsize=(10,6))
plt.barh(_10_shortest.index, _10_shortest["Duration (minutes)"], color="lightcoral")
plt.xlabel("Runtime (minutes)")
plt.ylabel("Movie Title")
plt.title("Top 10 Longest Movies")
plt.gca().invert_yaxis()  
plt.xticks(np.arange(0, _10_shortest["Duration (minutes)"].max() + 30, 5)) 
plt.show()
```

```python
#   ix) Enlisting all genres
all_genres = movies.groupby("Genre").size().reset_index(name="Movies")
print(all_genres)
```

```python
# presenting genres
plt.figure(figsize=(10,6))
plt.bar(all_genres["Genre"], all_genres["Movies"], color="violet")
plt.xlabel("Genre")
plt.ylabel("Number of Movies")
plt.title("Number of Movies per Genre")
plt.xticks(rotation=40, ha='right') 
plt.yticks(np.arange(0,all_genres["Movies"].max()+10,10))
plt.show()
```

```python
#   x)highest rating of each genre
highest_rating = movies.groupby("Genre")["IMDb Rating"].max().reset_index(name="Highest rating")
print(highest_rating)
```

```python
#   xi) title of the highest rated movie in each genre
idx_h = movies.groupby("Genre")["IMDb Rating"].idxmax()
highest_rated_per_genre = movies.loc[idx_h, ['Genre', 'IMDb Rating']]
highest_rated_per_genre = highest_rated_per_genre.drop_duplicates(subset="Genre")
print(highest_rated_per_genre)
```

```python
# Sort by rating for better visual
highest_rated_sorted = highest_rated_per_genre.sort_values("IMDb Rating", ascending=False)

plt.figure(figsize=(10,6))
bars = plt.barh(highest_rated_sorted["Genre"], highest_rated_sorted["IMDb Rating"], color="green")
plt.xlabel("IMDb Rating")
plt.xticks(np.arange(0,10.5,0.5))
plt.ylabel("Genre")
plt.title("Highest Rated Movie per Genre")
plt.gca().invert_yaxis()

# Add rating + movie name labels
for i, bar in enumerate(bars):
    rating = round(bar.get_width(), 1)
    movie_name = highest_rated_sorted.index[i]  # since title is your index
    plt.text(bar.get_width() - 5,
             bar.get_y() + bar.get_height()/2,
             f"{rating} ({movie_name})",
             va='center',
             color='white')

plt.show()
```

```python
#   xii) lowest rating of each genre
lowest_rating = movies.groupby("Genre")["IMDb Rating"].min().reset_index(name="Lowest rating")
print(lowest_rating)
```

```python
#   xiii) title of the lowest rated movie in each genre
idx_l = movies.groupby("Genre")["IMDb Rating"].idxmin()
lowest_rated_per_genre = movies.loc[idx_l, ['Genre', 'IMDb Rating']]
unique_lowest_rated_per_genre = lowest_rated_per_genre.drop_duplicates(subset="Genre")
print(unique_lowest_rated_per_genre)
```

```python
# Sort by rating for better visual
lowest_rated_sorted = lowest_rated_per_genre.sort_values("IMDb Rating", ascending=False)

plt.figure(figsize=(10,6))
bars = plt.barh(lowest_rated_sorted["Genre"], lowest_rated_sorted["IMDb Rating"], color="green")
plt.xlabel("IMDb Rating")
plt.xticks(np.arange(0,10.5,0.5))
plt.ylabel("Genre")
plt.title("Lowest Rated Movie per Genre")
plt.gca().invert_yaxis()

# Add rating + movie name labels
for i, bar in enumerate(bars):
    rating = round(bar.get_width(), 1)
    movie_name = lowest_rated_sorted.index[i]  # since title is your index
    plt.text(bar.get_width() - 2.5,
             bar.get_y() + bar.get_height()/2,
             f"{rating} ({movie_name})",
             va='center',
             color='white')

plt.show()
```

```python
#   xiv) average rating of each genre
average_rating_in_genre = movies.groupby("Genre")["IMDb Rating"].mean().reset_index(name="Average rating")
print(average_rating_in_genre)
```

```python
#   xv)Number of movies of each certificate
all_certificates = movies.groupby("Certificates").size().reset_index(name="Movie count")
all_certificates
```

```python
plt.figure(figsize=(10, 6))
plt.bar(all_certificates["Certificates"], all_certificates["Movie count"], color="pink")
plt.xticks(rotation=45, ha="right")
plt.yticks(np.arange(0,all_certificates["Movie count"].max()+5,10))
plt.title("Number of Movies by Certificate")
plt.xlabel("Certificate")
plt.ylabel("Movie Count")
plt.tight_layout()
plt.show()
```

```python
#   xvi) best movie of each certificate
idx_h1 = movies.groupby("Certificates")["IMDb Rating"].idxmax()
best_of_certificate = movies.loc[idx_h1,["IMDb Rating","Genre","Certificates"]]
best_of_certificate = best_of_certificate.drop_duplicates(subset="Certificates")
print(best_of_certificate)
```

```python
# Normalize ratings for color mapping
norm = plt.Normalize(best_of_certificate["IMDb Rating"].min(), 
                     best_of_certificate["IMDb Rating"].max())
colors = plt.cm.viridis(norm(best_of_certificate["IMDb Rating"]))

plt.figure(figsize=(10, 6))

# Create bars with colormap
bars = plt.bar(
    best_of_certificate["Certificates"],
    best_of_certificate["IMDb Rating"],
    color=colors
)

# Add labels above bars
for i, v in enumerate(best_of_certificate["IMDb Rating"]):
    plt.text(i, v + 0.05, f"{v:.1f}", ha='center', fontsize=9)

# Labels & title
plt.xlabel("Certificate")
plt.ylabel("IMDb Rating")
plt.title("Highest Rated Movie for Each Certificate")

plt.show()
```

```python
#   xvi) best movie of each certificate
idx_l1 = movies.groupby("Certificates")["IMDb Rating"].idxmin()
worst_of_certificate = movies.loc[idx_l1,["IMDb Rating","Genre","Certificates"]]
unique_W_O_C = worst_of_certificate.drop_duplicates(subset="Certificates")
print(unique_W_O_C)
```

```python
#   xvii) year wise distribution of movies
year_wise = movies.groupby("Year").size().reset_index(name="Movies Realeased")
year_wise
```

```python
plt.figure(figsize=(12, 6))
plt.plot(year_wise["Year"], year_wise["Movies Realeased"], marker='*', color="blue", linewidth=1)

# Labels and title
plt.xlabel("Year")
plt.ylabel("Number of Movies Released")
plt.title("Movies Released per Year")
plt.grid(True, linestyle="--", alpha=0.6)

plt.show()
```

```python
#   xviii) best movie years
high_movie_years = year_wise.sort_values("Movies Realeased",ascending=False,ignore_index=True)
high_movie_years.head(10)
```

```python
#   xix) highest rated film in each year
idx_h2 = movies.groupby("Year")["IMDb Rating"].idxmax()
highest_rated_per_year = movies.loc[idx_h2,["Year","IMDb Rating","Genre","Certificates"]]
highest_rated_per_year = highest_rated_per_year.drop_duplicates(subset="Year")
highest_rated_per_year
```

```python
#   xx) enlisting all works of the directors
all_movies_director = movies.groupby("Director").size().reset_index(name="Number of Movies")
all_movies_director = all_movies_director.sort_values("Number of Movies",ascending=False).reset_index()
all_movies_director
```

```python
#   xxi) best directors w.r.t average movie rating
best_director = movies.groupby("Director")["IMDb Rating"].mean().reset_index(name="Average rating").sort_values(by="Average rating", ascending=False)
best_director.head(10)
```

```python
#   xxii) movies of the best director
top_directors = best_director.iloc[0:11]["Director"]
best_director_movies = movies[movies["Director"].isin(top_directors)].sort_values(by="IMDb Rating",ascending=False)
best_director_movies
```
